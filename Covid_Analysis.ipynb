```python
import pandas as pd
import io

dataset = pd.read_csv(io.BytesIO(uploaded['covid19_Confirmed_dataset.csv']))
print(dataset)

```

```python
df=dataset.drop(["Lat","Long"], axis=1, inplace=True)
```

```python
#aggregate the rows by the country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
#Visualise data related to a country
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Afghanistan"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
#corona_dataset_aggregated.loc["China"].plot()
plt.legend()
```

```python
#Calculate the first derivate of the curve
corona_dataset_aggregated.loc["Spain"].diff().plot()
```

```python
#Maximum infection rate
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    # Convert the column to numeric and handle missing values
    diff_values = pd.to_numeric(corona_dataset_aggregated.loc[c], errors='coerce').diff().fillna(0)

    # Append the maximum infection rate to the list
    max_infection_rates.append(diff_values.max())

# Add the new column to the DataFrame
corona_dataset_aggregated["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated
```

```python
#Create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```
